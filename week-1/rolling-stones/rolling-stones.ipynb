
### Functions to build-out:

* **Searching functions**
  * Find by name - Takes in a string that represents the name of an album. Should return a dictionary with the correct album, or return `None`.
  * Find by rank - Takes in a number that represents the rank in the list of top albums and returns the album with that rank. If there is no album with that rank, it returns `None`.
  * Find by year - Takes in a number for the year in which an album was released and returns a list of albums that were released in that year. If there are no albums released in the given year, it returns an empty list.
  * Find by years - Takes in a start year and end year. Returns a list of all albums that were released on or between the start and end years. If no albums are found for those years, then an empty list is returned. 
  * Find by ranks - Takes in a start rank and end rank. Returns a list of albums that are ranked between the start and end ranks. If no albums are found for those ranks, then an empty list is returned.
* **All functions**
  * All titles - Returns a list of titles for each album.
  * All artists - Returns a list of artist names for each album.
* **Questions to answer / functions**
  * Artists with the most albums - Returns the artist with the highest amount of albums on the list of top albums 
  * Most popular word - Returns the word used most in amongst all album titles
  * Histogram of albums by decade - Returns a histogram with each decade pointing to the number of albums released during that decade.
  * Histogram by genre - Returns a histogram with each genre pointing to the number of albums that are categorized as being in that genre.

````python
def find_by_name(album_name):
    for album in albums:
        if album['album'] == album_name:
            return album or None
````

````python
find_by_name('Revolver')
````

````python
def find_by_rank(album_rank):
    for album in albums:
        if int(album['number']) == album_rank:
            return album or None
````

````python
find_by_rank(5)
````

````python
def find_by_year(album_year):
    released = []
    for album in albums:
        if int(album['year']) == album_year:
            released.append(album)
    return released
````

````python
# find by years takes 2 different years and returns all albums released on or between both years
````

````python
def find_by_years(start, end):
    released = []
    for album in albums:
        if int(album['year']) >= start and int(album['year']) <= end:
            released.append(album)
    return released
````

````python
len(find_by_years(2000, 2100))
````

````python
def find_by_ranks(start, end):
    released = []
    for album in albums:
        if int(album['number']) >= start and int(album['number']) <= end:
            released.append(album)
    return released
````

````python
def all_titles(albums):
    all_titles_list = []
    for album in albums:
        all_titles_list.append(album['album'])
    return all_titles_list
````

````python
all_titles(albums)[:5]
````

````python
def all_artists(albums):
    all_artists_list = []
    for album in albums:
        all_artists_list.append(album['artist'])
    return all_artists_list
````

````python
all_artists(albums)[:5]
````

````python
def artist_most_albums(albums):
    # create a list of all artists without duplicates
    artists = list(set(all_artists(albums)))
    
    # iterate through the set list to count how many times each distinct artist is in the FULL list of artists
    counter = [all_artists(albums).count(artist) for artist in artists]

    # find index of the max and return the corresponding artist name from the other list
#     return artists[counter.index(max(counter))]
    counted_dict = dict(zip(artists, counter))
    return sorted(counted_dict, key = counted_dict.__getitem__, reverse = True)[:5]
````

````python
artist_most_albums(albums)
````

````python
def most_pop_word(albums):
    # iterate through list of album titles
    titles = all_titles(albums)
    words = []
    for word in titles:
        words.extend(word.lower().split())
    word_count = Counter(words)
    return word_count.most_common(5)
    # .split each title by word
    # create set of words to loop through
    # count through set vs full list
    # return most popular word
````
